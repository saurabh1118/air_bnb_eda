```python
data = pd.read_csv('new_york_airbnb.csv', encoding_errors='ignore')
```

# **Task 4. Data Clening**

```python
data.isnull().sum()

#dropping all missing values rows
data.dropna(inplace=True)

#data.fillna()
data.isnull().sum()
```

```python
#Dealing with Duplicates rows
data.duplicated().sum()

#deleting all duplicates rows
#data[data.duplicate()]
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
#type casting
#changing data types

data.dtypes

data['id'] = data['id'].astype(object)
data.dtypes

data['host_id']= data['host_id'].astype(object)
data.dtypes
```

# **Task 5. Data Analysis**

# **Univariate Analysis**

```python
#identify outliers in price

df = data[data['price']<1500]
sns.boxplot(data=df, x='price')
```

```python
#Price distribution

plt.figure(figsize=(8,5))
sns.histplot(data=df,x='price',bins=100)
plt.title('Price Distribution')
plt.ylabel("Frequency")
plt.show()
```

```python
df.dtypes
```

```python
#Price Distribution
plt.figure(figsize=(6,3))
sns.histplot(data=df,x='availability_365')
plt.title('Availability_365 Distribution')
plt.ylabel("Frequency")
plt.show()
```

```python
data.dtypes
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

# **Feature Engineering**

```python
# average price per bed
df.groupby(by='neighbourhood_group')['price'].mean()
```

```python
#['price per bed]
df['price per bed']=df['price']/df['beds']
df.head()
```

```python
df.head()
```

**BI Variable Analysis :** One variable dependency in another variable

```python
df.columns
```

```python
#price dependency on neighourhood

sns.barplot(data=df,x='neighbourhood_group',y='price',hue='room_type')
```

```python
#Number of reviews and price rel
plt.figure(figsize=(8,5))
plt.title("Locality and Review dependency")
sns.scatterplot(data=df,x='number_of_reviews',y='price',hue='neighbourhood_group')
plt.show()
```

```python
df.dtypes
```

```python
sns.pairplot(data=df,vars=['price','minimum_nights','number_of_reviews',
'availability_365'],hue='room_type')
```

```python
#Geographical Distribution Of AirBnB listing
plt.figure(figsize=(10,7))
sns.scatterplot(data=df,x='longitude',y='latitude',hue='room_type')
plt.title('Geographical Distribution Of AirBnB listing')
plt.show()
```

```python
df.dtypes
```

```python
#heat map - Correlation of one variable with others for numerical column

corr= df[['latitude','longitude','price','minimum_nights','number_of_reviews',
          'reviews_per_month','availability_365','beds']].corr()

corr

plt.figure(figsize=(8,6))
sns.heatmap(corr,annot=True)
```
